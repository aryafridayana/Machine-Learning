```python
import pandas as pd

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/insurance.csv')

# Menampilkan beberapa baris pertama dari DataFrame
df.head()
```

# Feature Statistic

```python
# Menampilkan statistik fitur
print("Feature Statistic:")
df.describe()
```

# Box Plot

```python
df. info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Memisahkan fitur numerik
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Membuat box plot untuk setiap fitur numerik
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.set_palette("pink")
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

# Scatter Plot

```python
# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(df.select_dtypes(include=['float64', 'int64']), kind='scatter')
plt.suptitle('Scatter Plot untuk Fitur Numerik', fontsize=16)
plt.show()
```

# Heat Map

```python
# Menghitung korelasi antar fitur numerik
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Heatmap dari Korelasi Fitur Numerik', fontsize=16)
plt.show()
```

# Violin Plot


```python
# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Membuat violin plot untuk setiap fitur numerik
plt.figure(figsize=(12, 8))
sns.violinplot(data=df.select_dtypes(include=['float64', 'int64']), palette="pastel", inner="points")
plt.title('Violin Plot untuk Fitur Numerik', fontsize=16)
plt.xticks(rotation=45)
plt.show()
```

# Histogram

```python
# Mengatur ukuran gambar
plt.figure(figsize=(12, 8))

# Menampilkan histogram untuk setiap fitur numerik
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.hist(df[column], bins=20, alpha=0.7, label=column)

# Menambahkan label dan judul
plt.xlabel('Nilai Fitur')
plt.ylabel('Frekuensi')
plt.title('Histogram untuk Fitur Numerik')
plt.legend()

# Menampilkan plot
plt.show()
```

# Bar Plot

```python
# Menghitung jumlah polis asuransi untuk setiap region
insurance_count = df['region'].value_counts()

# Membuat bar plot
plt.figure(figsize=(8, 6))
insurance_count.plot(kind='bar', color='skyblue')
plt.title('Jumlah Polis Asuransi Berdasarkan Region')
plt.xlabel('Region')
plt.ylabel('Jumlah Polis Asuransi')
plt.xticks(rotation=0)  # Mengatur label sumbu x tidak miring
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Menambahkan grid di sumbu y
plt.show()
```

# Pie Chart

```python
# Menghitung jumlah polis asuransi untuk setiap region
insurance_count = df['region'].value_counts()

# Membuat pie chart
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(insurance_count, labels=insurance_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart dari Jumlah Polis Asuransi Berdasarkan Region')
plt.axis('equal')  # Agar lingkaran terlihat sebagai lingkaran

# Menghitung rata-rata biaya asuransi berdasarkan umur
avg_charges_by_age = df.groupby('age')['charges'].mean()
```

# Line Chart

```python
# Membuat line chart
plt.subplot(1, 2, 2)
plt.plot(avg_charges_by_age.index, avg_charges_by_age.values, marker='o', color='green', linestyle='-')
plt.title('Rata-Rata Biaya Asuransi Berdasarkan Umur')
plt.xlabel('Umur')
plt.ylabel('Rata-Rata Biaya Asuransi')
plt.grid(True)

plt.tight_layout()  # Agar subplot tidak tumpang tindih
plt.show()
```
