```python
import pandas as pd

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/diamonds.csv')

# Menampilkan beberapa baris pertama dari DataFrame
df.head()
```

# Feature Statistic


```python
# Menampilkan statistik fitur
print("Feature Statistic:")
df.describe()
```

# Box Plot

```python
df. info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Memisahkan fitur numerik
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Membuat box plot untuk setiap fitur numerik
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.set_palette("pink")
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

# Scatter Plot

```python
# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(df.select_dtypes(include=['float64', 'int64']), kind='scatter')
plt.suptitle('Scatter Plot untuk Fitur Numerik', fontsize=16)
plt.show()
```

# Heat Map

```python
# Menghitung korelasi antar fitur numerik
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Heatmap dari Korelasi Fitur Numerik', fontsize=16)
plt.show()
```
